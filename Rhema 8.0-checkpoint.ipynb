# HOW TO CHANGE FONT SIZE AND STYLE

```python
from tkinter import *
def salary_():
    amount=float(amt.get())
    increment=amount*0.25
    new_salary=round(amount+increment,2)
    outcome.config(text=f"New Salary:${new_salary}")
    
naruto=Tk(className=" SALARY")
naruto.config(bg="green")

frame=Frame(naruto)
frame.pack()

Label(frame,
      text="Enter Salary",
      bg="red",
      fg="white",
      font=("Helvatical bold",18)).pack(side=LEFT)

amt=Entry(frame)
amt.pack()

Button(naruto,
       text="Enter",
       font=("Times",18),
       bg="orange",
       fg="white",
       command=salary_).pack()

outcome=Label(naruto)
outcome.pack()

naruto.mainloop()
# font=font is used to change the text style and font
```

# HOW TO INCLUDE INTERNAL AND EXTERNAL PADDING

```python
from tkinter import *
anime=Tk()
anime.title("HELLO")
anime.config(bg="light blue")
#anime.geometry("400400")
Label(anime,
      text="What's your name",
      font=("Ariel",16),
      bg="white",
      fg="black").pack()

Entry(anime).pack(padx=(100,0))
anime.mainloop()
```

```python
#mod1
from tkinter import *
def name():
    amq=entry.get()
    out.config(text=f"Hello {amq}",
               bg="pink",
               fg="white",
               font=("Ariel black",16))
    
anime=Tk()
anime.title("HELLO")
anime.config(bg="light blue")
#anime.geometry("400*400")
Label(anime,
      text="What's your name",
      font=("Ariel",16),
      bg="white",
      fg="black").pack(padx=(59,12))#

entry=Entry(anime)
entry.pack(pady=60)#

Button(anime,
       text="Enter",
       font=("Times",16),
       bg="sky blue",
       fg="black",
       command=name).pack(pady=100)#

out=Label(anime)
out.pack()

anime.mainloop()
#padx:placing the widget in the window on x axis
#pady: placing the widget in the window on y axis
```

# RADIO_BUTTON WIDGET

```python
from tkinter import *
fly=Tk(className=" RADIO BUTTON")
fly.config(bg="light pink")

Label(fly,
      text="What's your favourite anime",
      font=("ariel black",15),
      bg="light blue").pack()

Radiobutton(fly,
            text="Naruto",
            activebackground="black",
            activeforeground="red",
            bg="orange",
            bd=12,
            cursor="target").pack()

Radiobutton(fly,
            text="Bleach",
            activebackground="black",
            activeforeground="red",
            bg="black",
            bd=12,fg="white",
            cursor="target").pack()

Radiobutton(fly,
            text="One Piece",
            activebackground="black",
            activeforeground="red",
            bg="red",
            bd=12,
            cursor="target").pack()

fly.mainloop()
# activebackground= The active background is used to change background colour when the mouse is over the radio button
# activeforeground= The active foreground is used to change background colour when the mouse is over the radio button
# bd= the bd is used to change the border width in pixels. defult is 2
#cursor= the mouse cursor will change to that pattern when it is over the radio button
```

```python
# mod 1
from tkinter import *
fly=Tk(className=" RADIO BUTTON")
fly.config(bg="light pink")

Label(fly,
      text="What's your favourite Series",
      font=("ariel black",15),
      bg="light blue").pack()

Radiobutton(fly,
            text="Marlin",
            fg="blue",
            font="Castellar",
            height=3,
            width=10,
            relief="solid").pack()

Radiobutton(fly,
            text="Flash",
            fg="red",
            font="Castellar",
            height=3,
            width=10,
            relief="ridge").pack()

Radiobutton(fly,
            text="GOT",
            fg="brown",
            font="Castellar",
            height=3,
            width=10,
            relief="solid").pack()

fly.mainloop()
#height - The height is used to change the height on the radio button. in text lines (text buttons) or pixels (images).
#width - The width is used to change the width on the radio button in letters.
#Relief - Relief is used to specify the type of the border. The values are SUNKEN, RAISED, GROOVE, and RIDGE,SOLID.
```

```python
# mod 2
from tkinter import *
fly=Tk(className=" RADIO BUTTON")
fly.config(bg="light pink")

Label(fly,
      text="What's your favourite Chocolate bar",
      font=("ariel black",15),
      bg="light blue").pack()

Radiobutton(fly,
            text="Snickers",
            state=DISABLED,
            justify=RIGHT,
            padx=11,
            pady=10).pack()

Radiobutton(fly,
            text="Mars",
            state=DISABLED,
            justify=LEFT,
            padx=11,
            pady=10).pack()

Radiobutton(fly,
            text="M&M",
            state=DISABLED,
            justify=CENTER,
            padx=11,
            pady=10).pack()

fly.mainloop()
#State - The state is used to set this option to DISABLED to gray out the radio button and make it unresponsive.
#Justify - The justify is used to change the arrangement order LEFT, CENTER, RIGHT.
#Padx - The padx is used to change the additional padding left and right of the text.
#Pady - The pady is used to changing the additional padding above and below the text
```

# CHECKBOX WIDGET

```python
from tkinter import *
 
root =Tk()
 
c = Checkbutton(root, text = "Python")
c.pack()
 
c1 = Checkbutton(root, text = "C++")
c1.pack()
 
c2 = Checkbutton(root, text = "C")
c2.pack()
 
root.geometry("400x400+120+120")
root.mainloop()
```
